```julia
using PyPlot
```

```julia
include("sneasy.jl")
```

```julia
model = getsneasy()
```

```julia
run(566,model)
```

```julia
plot(collect(1850:1850+566-1), model[2].v.temp)
title("temperature")
xlabel("year")
ylabel("°C")
```

```julia
plot(collect(1850:1850+566-1), model[1].v.rf)
title("radiative forcing")
xlabel("year")
ylabel("W/m^2")
```

```julia
plot(collect(1850:1850+566), model[3].v.atmco2)
title("CO2 concentrations")
xlabel("year")
ylabel("ppm")
```
